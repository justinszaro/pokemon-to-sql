```python
data = pd.read_csv('pokedex.csv')
```

```python
data.head(10)
```

```python
def getRegionName(generation):
    region = {1: 'Kanto', 2: 'Johto', 3: 'Hoenn', 4: 'Sinnoh', 5: 'Unova', 6: 'Kalos', 7: 'Alola', 8: 'Galar'}
    return region[generation]
```

```python
data['region_name'] = data['generation'].apply(getRegionName)
```
